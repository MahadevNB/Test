```python
from pymongo import MongoClient

# Local MongoDB (make sure MongoDB service is running)
client = MongoClient("mongodb://localhost:27017/")

# Or MongoDB Atlas (replace with your connection string)
# client = MongoClient("mongodb+srv://<username>:<password>@cluster0.mongodb.net/")
```

```python
# List all databases
print(client.firstdb)
```

```python
db = client["firstdb"]          # my dummy data base name
collection = db["firstcollection"]    # my dummy collection name
```

```python
# Get all documents
data = list(collection.find())

# Preview first 5 documents
for doc in data[:5]:
    print(doc)
    
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head()
```
