```python
MONGO_URI = "mongodb://localhost:27017/"
DB_NAME   = "EncoreDB"

GA4_MEASUREMENT_ID = "G-6ZVFSRTJL3"
GA4_API_SECRET     = "ji7GndWjReqJXpTEdH_FjQ"

GA4_ENDPOINT = (
    f"https://www.google-analytics.com/mp/collect"
    f"?measurement_id={GA4_MEASUREMENT_ID}"
    f"&api_secret={GA4_API_SECRET}"
)

print("✅ Config ready")
```

```python
import requests
import threading
from pymongo import MongoClient

def send_to_ga4(event_name, params):
    clean_params = {k: str(v) for k, v in params.items() if v is not None}
    payload = {
        "client_id": "mongo-realtime-sync",
        "events": [{"name": event_name, "params": clean_params}]
    }
    try:
        res = requests.post(GA4_ENDPOINT, json=payload, timeout=5)
        if res.status_code == 204:
            print(f"✅ Sent '{event_name}' to GA4")
        else:
            print(f"⚠️ GA4 status: {res.status_code}")
    except Exception as e:
        print(f"❌ Error: {e}")

def make_event_name(col, op):
    return f"{col.replace('-','_').lower()}_{op}"

def flatten_doc(doc, prefix=""):
    flat = {}
    for k, v in doc.items():
        key = f"{prefix}_{k}" if prefix else k
        if isinstance(v, dict):
            flat.update(flatten_doc(v, prefix=key))
        else:
            flat[key] = str(v)
    return flat

def watch_collection(col_name):
    client = MongoClient(MONGO_URI)
    col = client[DB_NAME][col_name]
    print(f"👀 Watching: {col_name}")
    pipeline = [{"$match": {"operationType": {"$in": ["insert","update","delete"]}}}]
    try:
        with col.watch(pipeline) as stream:
            for change in stream:
                op = change["operationType"]
                event = make_event_name(col_name, op)
                if op == "insert":
                    doc = change.get("fullDocument", {})
                    doc.pop("_id", None)
                    params = flatten_doc(doc)
                elif op == "update":
                    params = flatten_doc(change.get("updateDescription",{}).get("updatedFields",{}))
                else:
                    params = {}
                params["doc_id"] = str(change["documentKey"]["_id"])
                print(f"📦 {col_name}: {op}")
                send_to_ga4(event, params)
    except Exception as e:
        print(f"❌ Error watching '{col_name}': {e}")

print("✅ Functions ready")
```

```python
client = MongoClient(MONGO_URI)
collections = client[DB_NAME].list_collection_names()
client.close()

print(f"🚀 Found {len(collections)} collections: {collections}\n")

for col in collections:
    threading.Thread(target=watch_collection, args=(col,), daemon=True).start()

print("\n✅ Watching all collections! Keep this cell running.")
```
