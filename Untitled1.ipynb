```python
from flask import Flask, jsonify, send_from_directory
from flask_cors import CORS
from pymongo import MongoClient
import pandas as pd
import os

app = Flask(__name__, static_folder=".")
CORS(app)

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["EncoreDB"]




@app.route("/")
def index():
    return send_from_directory(".", "dashboard.html")

@app.route("/api/top-customers")
def top_customers():
    df = get_orders_df()
    result = (
        df.groupby([
            'order_customer_name',
            'order_customer_UID',
            'order_customer_contact_email',
            'order_customer_contact_phone'
        ])['order_Overall_Price']
        .sum()
        .reset_index()
        .sort_values(by='order_Overall_Price', ascending=False)
        .head(10)
    )
    result['order_Overall_Price'] = result['order_Overall_Price'].round(2)
    return jsonify(result.to_dict(orient="records"))

@app.route("/api/summary")
def summary():
    df = get_orders_df()
    total_revenue = df['order_Overall_Price'].sum()
    total_orders = len(df)
    unique_customers = df['order_customer_UID'].nunique() if 'order_customer_UID' in df.columns else 0
    avg_order = df['order_Overall_Price'].mean()
    return jsonify({
        "total_revenue": round(float(total_revenue), 2),
        "total_orders": int(total_orders),
        "unique_customers": int(unique_customers),
        "avg_order_value": round(float(avg_order), 2)
    })

@app.route("/api/chart-data")
def chart_data():
    df = get_orders_df()
    result = (
        df.groupby('order_customer_name')['order_Overall_Price']
        .sum()
        .reset_index()
        .sort_values(by='order_Overall_Price', ascending=False)
        .head(10)
    )
    return jsonify({
        "labels": result['order_customer_name'].tolist(),
        "values": result['order_Overall_Price'].round(2).tolist()
    })

if __name__ == "__main__":
    print("🚀 EncoreDB Dashboard running at http://localhost:5000")
    app.run(debug=True, port=5000)
```
