```python
import pymongo
import pandas as pd

# 1. Connect to MongoDB (adjust URI if needed)
client = pymongo.MongoClient("mongodb://localhost:27017/")

# 2. Select your database
db = client["EncoreDB"]

# 3. List of collections you mentioned
collections = [
    "productfolds",
    "productgirths",
    "quotationmasters",
    "quotedesigndefectlogs",
    "quoteitems",
    "quotesitemsmanuals",
    "rackings",
    "roles",
    "rolespermissioninfos",
    "runsdriverorders",
    "runsdrivers",
    "splitorders",
    "suppliers",
    "template_libraries",
    "templates",
    "truck_maintainances",
    "trucks",
    "userorders",
    "users"
]

# 4. Load each collection into a DataFrame
dfs = {}
for coll in collections:
    cursor = db[coll].find()
    df = pd.DataFrame(list(cursor))
    dfs[coll] = df

# 5. Example: preview one DataFrame
print(dfs["users"].head())
```
